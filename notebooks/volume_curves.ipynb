```python
import os 
import numpy as np
import matplotlib.pyplot as plt
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
```

```python
# for some reason the flow environment is trading faster 
# doesn't make much sense 

tag = None 
tag = 'deterministic_action'
plt.figure(figsize=(10, 6))
fig, axs = plt.subplots(2, 3, figsize=(20, 6))
time = np.arange(0, 155, 15)
for i, market in enumerate(['noise', 'flow', 'strategic']):
    for rl in ['log_normal', 'dirichlet']:
        ax = axs[0, i] if rl == 'log_normal' else axs[1, i]
        for lots in [20, 60]:
            # 10000 or 1000 doesnt make a difference
            if tag:
                path = f'{parent_dir}/volume_curves/{market}_{lots}_samples_1000_{rl}_{tag}.npz'
                # print('path with tag:', path)
                # path = '/u/weim/lob/volusme_curves/noise_20_samples_1000_log_normal_deterministic_action.npz'
                print('path with tag:', path)
            else:
                path = f'{parent_dir}/volume_curves/{market}_{lots}_samples_1000_{rl}.npz'
            data = np.load(path)['volume_curves']
            mean_volume = np.mean(data, axis=0)
            std_volume = np.std(data, axis=0)
            ax.plot(time, mean_volume, label=f'Lots: {lots}')
            ax.fill_between(time, mean_volume - std_volume, mean_volume + std_volume, alpha=0.2)
            ax.set_xticks(time)
            ax.set_xlim(0, 150)
            ax.set_ylim(0, 60)
        if i == 0:
            ax.legend()
            ax.set_ylabel('Volume')
        ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.set_title(f'{market.capitalize()}')
plt.show()
```

```python
## create plots for multiple for each market seperately 
# just plot this for fhe logistic normal 

# import seaborn as sns
# only do the logistic normal distribution for now 

lots_list = [20, 60]
textwidth = 6.75
textwidth = 12 
textheight = 14
scale = 2
lots_list = [10, 20, 60]
rl = 'log_normal'
rl = 'dirichlet'
tag = 'deterministic_action'

for i, env in enumerate(['noise', 'flow', 'strategic']):
    print('#############')
    print('generating plots for ', env)
    fig, axs = plt.subplots(2, 1, figsize=(scale*0.3*textwidth, scale*0.3*textheight), dpi=300)
    for j, rl in enumerate(['log_normal', 'dirichlet']): 
        if rl == 'log_normal':
            colors = ['#006400', "#63E70B"]  # dark green, darker green (second not lighter than first)
        else:
            colors = ['#8b0000', "#f04e4e"]  # dark red, lighter red
        ax = axs[j] 
        ax.set_prop_cycle(color=colors)
        for lots in [60, 20]:
            if tag:
                path = f'{parent_dir}/volume_curves/{env}_{lots}_samples_1000_{rl}_{tag}.npz'
            else:
                path = f'{parent_dir}/volume_curves/{env}_{lots}_samples_1000_{rl}.npz'
            print('loading data from ', path)
            data = np.load(path)['volume_curves']
            mean_volume = np.mean(data, axis=0)
            std_volume = np.std(data, axis=0)
            ax.plot(time, mean_volume, label=f'Lots: {lots}', linewidth=3)
            ax.fill_between(time, mean_volume - std_volume, mean_volume + std_volume, alpha=0.2)
            ax.set_xticks(np.arange(0, 155, 15))
            ax.set_xlim(0, 150)
            if env == 'noise':   
                ax.legend(prop={'size': 16}, loc='upper right')
            if rl == 'dirichlet':
                ax.set_xlabel('Time', fontsize=20)
            if env == 'noise':
                ax.set_ylabel('Inventory', fontsize=20)
            ax.set
            tick_size = 14
            ax.set_xlim(0, 150)
            ax.set_ylim(0, 60)
            ax.tick_params(axis='x', labelsize=tick_size)
            ax.tick_params(axis='y', labelsize=tick_size)
            # ax.grid(axis='y', linestyle='--', alpha=0.8)
            ax.grid(alpha=0.8)
            # ax.set_yticks(np.arange(-2.5, 0.5, 0.5))
    fig.set_dpi(150)
    plt.tight_layout()
    if tag:
        plt.savefig(f'{parent_dir}/plots/vcurves_{env}_{tag}.pdf')
    else:
        plt.savefig(f'{parent_dir}/plots/vcurves_{env}.pdf')
    plt.show()


```
